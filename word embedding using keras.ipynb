# word embedding technique using embedding layer in keras

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent=['the glass of milk',
     'the glass of juice',
     'the cup of tea',
     'I am a good boy',
     'I am a good developer',
     'Understand the meaning of words',
     'Your videos are good',]
```

```python
sent
```

Convert these sentences to one hot

```python
vec_size=10000 # dict size
```

```python
onehot_repr=[one_hot(words,vec_size) for words in sent]
```

```python
onehot_repr
```

word embedding representation

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
import numpy as np
```

```python
sent_size=8
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_size)
print(embedded_docs)
```

```python
dim=10 #feature size
```

```python
from keras.layers import Dense
from keras.layers import Flatten
```

```python
model=Sequential()
model.add(Embedding(vec_size,dim,input_length=sent_size))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
